```python
import matplotlib.pyplot as plt
import numpy as np

#from 0 to 10 
x=np.arange(1,10,0.1)
y=np.sin(x)

plt.plot(x,y)

plt.title("Sine graph")
plt.xlabel("x axis")
plt.ylabel("y axis")
#graph printing
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.random.rand(50)
y=np.random.rand(50)
colors=np.random.rand(50)
#point chart datas 
sizes=np.random.randint(50,150,50)

plt.scatter(x,y,c=colors,s=sizes)

plt.title("point chart")
plt.xlabel("x axis")
plt.ylabel("y axis")

plt.show()
```

```python
import matplotlib.pyplot as p
import numpy as np

data = np.random.randn(1000)
#creating histogram chart
p.hist(data,bins=30)

#chart props
p.title("histogram")
p.xlabel("values")
p.ylabel("frequency")

p.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

#x has 5 element y should be increse 5 by 5
x=['A','B','C','D','F']
y=np.random.randint(1,10,5)

#column chart creation
plt.bar(x,y)

#chart props
plt.title("column chart")
plt.xlabel("values")
plt.ylabel("frequency")
plt.show()
```
